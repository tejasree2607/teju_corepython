```python
sns.barplot(data=tr,x='Pclass',y='Survived',palette='spring', errorbar=None)
plt.title("Survival rate by passenger class" )
plt.xlabel("Pclass")
plt.ylabel("Surviveal rate")
plt.show()
```

```python
sns.scatterplot(data=tr,x='Age',y='Fare', hue='Sex',palette='bwr')
plt.title("Comparing age and fare" )
plt.xlabel("age")
plt.ylabel("Fare")
plt.show()
```

```python
tem = tr[['Age', 'Fare', 'Pclass','Survived']]
sns.pairplot(tem, hue='Survived', diag_kind='kde')
plt.title("pair plot" )
plt.show()
```

```python
numtr=tr.select_dtypes(include=[float,int])
corrmat=numtr.corr()
sns.heatmap(corrmat,annot=True,cmap='RdPu',fmt=".2f")
plt.title("correlation matrix")
plt.show()
```
