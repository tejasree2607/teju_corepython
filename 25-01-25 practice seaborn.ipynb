```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips=sns.load_dataset("tips")
tips.head()
```

```python

sns.histplot(data=tips,x='total_bill',bins=20,color='yellow', edgecolor='orange',kde=True)
plt.title("histogram for distribution of total bill" )

plt.show()
```

```python
sns.kdeplot(data=tips,x='total_bill',fill=True,color='blue')
plt.title("kde for distribution of total bill" )
plt.show()
```

```python
sns.barplot(data=tips,x='day',y='total_bill',palette='Oranges')
plt.title("basic bar graph" )
plt.xlabel("Days")
plt.ylabel("bill")
plt.show()
```

```python
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex',palette='cividis')
plt.title("basicbox plot  graph" )
plt.xlabel("Days")
plt.ylabel("bill")
plt.show()
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='day',style='time',palette='deep')
plt.title("basicbox plot  graph" )
plt.xlabel("Days")
plt.ylabel("bill")
plt.show()
```

```python
sns.pairplot(data=tips,hue='sex',palette='husl')
plt.title("basic pairplot  graph" )
plt.xlabel("Days")
plt.ylabel("bill")
plt.show()
```

```python
sns.lineplot(data=tips,x='size',y='tip',color='green', marker='o')
plt.title("basic line plot graph" )
plt.xlabel("Days")
plt.ylabel("bill")
plt.show()
```

```python
numtips=tips.select_dtypes(include=[float,int])
corrmat=numtips.corr()
sns.heatmap(corrmat,annot=True,cmap='BuPu',fmt=".2f")
plt.title("correlation matrix")
plt.show()
```
