```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("headbrain.csv")
```

```python
df
```

```python
x_df=df['Head Size(cm^3)']
y_df=df['Brain Weight(grams)']
```

```python
x_mean=np.mean(x_df)
y_mean=np.mean(y_df)
m=np.sum((x_df-x_mean)*(y_df-y_mean))/np.sum((x_df-x_mean)**2)
c=y_mean-(m*x_mean)
print(f'The equation for the dataset is y={m:.2f}x+{c:.2f}')
```

```python
pred=x_df*m+c
```

```python
#r square
ssr=np.sum((pred-y_mean)**2)
sst=np.sum((y_df-y_mean)**2)
rsq=1-(ssr/sst)
print(f"{rsq:.2f}")
```

```python
plt.figure(figsize=(10,8))
plt.scatter(x_df,y_df,color='blue' , label='Actual data')
plt.scatter(x_df,pred,color='orange' , label='Predicted data')
plt.plot(x_df,pred, color='green',label='Best fit line')
plt.legend()
plt.xlabel("Head Size")
plt.ylabel("Brain Weight")
plt.title("Simple Linear Regression")
plt.show()
```
