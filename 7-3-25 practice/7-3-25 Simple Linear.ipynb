```python
import numpy as np
import pandas as pd
df=pd.DataFrame({'x':[1,2,3,4,5],
                 'y':[3,4,2,4,5]})
x_mean=np.mean(df['x'])
y_mean=np.mean(df['y'])
m=np.sum((df['x']-x_mean)*(df['y']-y_mean))/np.sum((df['x']-x_mean)**2)
c=y_mean-(m*x_mean)
print(f'The equation for the dataset is y={m}x+{c}')
```

```python
pred=df['x']*m+c
pred
```

```python
#r square
ssr=np.sum((pred-y_mean)**2)
sst=np.sum((df['y']-y_mean)**2)
rsq=1-(ssr/sst)
print(f"{rsq:.2f}")
```
